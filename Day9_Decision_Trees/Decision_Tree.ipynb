```python
import pandas as pd
df = pd.read_csv("/content/salaries.csv")
df.head()
```

```python
inputs = df.drop(['salary_more_then_100k'], axis='columns')
inputs.head()
```

```python
target = df['salary_more_then_100k']
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
inputs
```

```python
inputs.drop(['company', 'job', 'degree'], axis='columns', inplace=True)
inputs
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
model = tree.DecisionTreeClassifier()
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, train_size=0.8, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[2,2,1]])
```

Exercise: Using titanic.csv, predict who would have survived

```python
df2 = pd.read_csv("/content/titanic.csv")
df2.head()
```

```python
df2.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns', inplace=True)
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2['Age'] = df2['Age'].fillna(df2['Age'].mean())
```

```python
df2.isnull().sum()
```

```python
inputs = df2.drop(['Survived'], axis='columns')
inputs.head()
```

```python
target = df2.Survived
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
```

```python
inputs['sex_n'] = le_sex.fit_transform(inputs['Sex'])
inputs.drop(['Sex'],axis='columns', inplace=True)
inputs.head()
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
titanic_model = tree.DecisionTreeClassifier()
titanic_model.fit(X_train, y_train)
```

```python
titanic_model.score(X_test, y_test)
```
